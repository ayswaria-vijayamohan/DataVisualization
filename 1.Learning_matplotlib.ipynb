# 1. Input Data

## 1.1. Hard Code x and y as a List

```python
x =[0,0.2,0.4,0.6,0.8,2.0]
y =[0,0.1,0.4,0.9,1.6,2.5]

plt.plot(x,y,scalex=False, scaley=True) # you can see the effect of scalex and scaley parameters on the plot
# when scalex is False, the x-axis is limited to 1, but scaley is not limited
# By default, scalex and scaley are both True
plt.show()
```

## 1.2. Range and Arange 

```python
import numpy as np

x = range(0, 11, 1) # here the interval is 1, it is a integer
y = np.arange(0,1.1, 0.1) # if you want to create an array of evenly spaced float values, use np.arange

plt.plot(x,y)
plt.show()
```

## 1.3. Mathematical Equations - we can see how to you use relations between x and y and generating the plot data

```python
import numpy as np

x = range(0, 11, 1) # here the interval is 1, it is a integer
y = [i**2 for i in x] # as you would have figured out, this is a parabola -> y=x^2


plt.plot(x,y)
plt.show()
```

## 1.4. Using available/inbuilt Math Functions

```python
import numpy as np

x = np.linspace(0, 10, 100) # when we use linspace, we get 100 points between 0 and 10, 
# for linspace we don't specify the interval but the number of points required between the start and end
y = np.cos(x) # thats a cosine wave

plt.plot(x,y)
plt.show()
```

# 2. Let's add more content to our Plot to make it readable! 

```python
import numpy as np

x = np.linspace(0, 10, 100) 
y = np.cos(x) # thats a cosine wave

plt.plot(x,y)

plt.title("Cosine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis([-1, 10, -2, 2])
plt.grid()
plt.axhline(0, color='green',linewidth=0.9, ls='--')
plt.axvline(0, color='red',linewidth=0.7, ls='--')

plt.show()
```

**Double the Fun!** Let's add a plot buddy to keep our cosine wave company! Two waves are better than one!


```python
import numpy as np

x = np.linspace(0, 10, 100) 
y = np.cos(x) # thats a cosine wave
y2 = np.sin(x) # thats a sine wave

plt.plot(x,y, label='Cosine') #mention the labels, so that they appear in the legend with the name given
plt.plot(x,y2, label='Sine')

plt.title("Cosine and Sine Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()

#  when there are more than one plot on same figure, its is better to use legend
plt.legend()
# if you dont mention it in the plot function, you can add it later in the legend as plt.legend(['Cosine', 'Sine'])
# but in this case you have to maintain the order of the plots, and that is difficult if there are multiple traces
plt.axhline(0, color='green',linewidth=0.9, ls='--')
plt.axvline(0, color='red',linewidth=0.7, ls='--')

plt.show()
```

# 3. Let's change few characteristics of the plot like colour marker and so on

```python
import numpy as np
import matplotlib.pyplot as plt

print('available styles are :', plt.style.available)
plt.style.use('fast')
x = np.linspace(0, 10, 100) 
y = np.cos(x) # thats a cosine wave

plt.plot(x,y, color='purple', linestyle='-', linewidth=0.5, marker='>', markersize=5, markerfacecolor='green', markeredgewidth=1, markeredgecolor='blue')

# - supported values for linestyle are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.title("Cosine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.axhline(0, color='green',linewidth=0.9, ls='--')
plt.axvline(0, color='red',linewidth=0.7, ls='--')

plt.show()
```


# 4. Another and probably the better approach to use matplotlib: Object-Oriented Matplotlib! 

## 4.1 Import

```python
import matplotlib.pyplot as plt
```

> Matplotlib is built with **figures** (the main tent) and **axes** (the performance stages)! 

```python

fig, axs = plt.subplots() # even though we used subplots we didnot mention the number of rows and columns, so in effect it is a single subplot
x = np.linspace(0, 10, 100)
y = np.cos(x)
y2 = np.sin(x)

axs.plot(x, y, label='Cosine') # both cos and sin are plotted on a single axes
axs.plot(x, y2, label='Sine')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 10)) # now the difference is nrows and ncols are specified, It creates 2 subplots. 2 rows and 1 column
x = np.linspace(0, 10, 100)
y = np.cos(x)
y2 = np.sin(x)

axs[0].plot(x, y, label='Cosine') # each cos and sin are plotted on separate axes
axs[1].plot(x, y2, label='Sine')
plt.show()
```

# 6. Why is choosing the right plot important?
>> Which plot to use when?


As data scientist, visualization hold immense power at different stages:
-- During the initial EDA phase:
- Data Quality Check  Detect missing values, outliers, and inconsistencies early.

-- During solution development:
- Model Debugging & Explainability  Visualize errors, biases, and feature impacts (e.g., SHAP, LIME).

- Feature Engineering Aid  Identify relationships, correlations, and transformations visually.

- Hypothesis Testing  Quickly validate assumptions and observe relationships before formal testing.

-- Result inference and stakeholder buy in:
- Collaboration & Communication  Provide a common visual language for technical and business teams.

-- Deployment:
- Monitoring & Drift Detection  Track model performance and detect data or concept drift post-deployment.
